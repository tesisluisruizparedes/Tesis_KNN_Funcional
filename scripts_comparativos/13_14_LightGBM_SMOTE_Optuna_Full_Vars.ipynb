```python
# 15_5: LightGBM + SMOTE + Optuna + Full Vars
# ==========================================

# ðŸš— Montar Drive
from google.colab import drive
drive.mount('/content/drive')
!pip install optuna lightgbm --quiet

# ðŸ”§ LibrerÃ­as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, gc
import optuna
from tqdm import tqdm
from lightgbm import LGBMClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, log_loss, roc_curve, precision_recall_curve,
    average_precision_score
)

# ðŸ“ Rutas
ruta_base = "/content/drive/MyDrive/Datos/6_Base_Modelos_Predictivos.parquet"
ruta_resumen = "/content/drive/MyDrive/Resultados/resultados_comparativos_modelos_turismo.csv"
ruta_roc = "/content/drive/MyDrive/Resultados/15_5_Curva_ROC_LightGBM_Optuna.png"
ruta_prc = "/content/drive/MyDrive/Resultados/15_5_Curva_PRC_LightGBM_Optuna.png"

# ðŸ“Š Cargar base
df = pd.read_parquet(ruta_base).drop_duplicates()
print(f"âœ”ï¸ Base cargada: {df.shape[0]:,} filas")

# ðŸŽ¯ Variables
y = df["RQ"]
X = df.drop(columns=["RQ", "NIT"])
X = pd.get_dummies(X, columns=["DEP", "CIIU_Letra", "AÃ±o"], drop_first=False)

# ðŸ” Infinitos â†’ grandes valores
X = X.replace([np.inf, -np.inf], 1e6)

# âš–ï¸ DistribuciÃ³n
print("\nðŸ“Š DistribuciÃ³n de clases:")
print(y.value_counts(normalize=True).rename(lambda x: f"RQ={x}"))

# ðŸ” Optuna
def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 500),
        "max_depth": trial.suggest_int("max_depth", 3, 12),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "random_state": 42
    }

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    f1_scores = []

    for train_idx, test_idx in skf.split(X, y):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        X_res, y_res = SMOTE(random_state=42).fit_resample(X_train, y_train)

        model = LGBMClassifier(**params)
        model.fit(X_res, y_res)
        y_pred = model.predict(X_test)
        f1_scores.append(f1_score(y_test, y_pred))

    return np.mean(f1_scores)

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=20)
mejores_params = study.best_params
print("ðŸŽ¯ Mejores parÃ¡metros:", mejores_params)

# ðŸ§ª ValidaciÃ³n cruzada final
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
resultados, loglosses, aucs, y_true_total, y_prob_total = [], [], [], [], []
roc_curves, prc_curves = [], []

for train_idx, test_idx in tqdm(skf.split(X, y), total=10, desc="LightGBM Optimizado CV"):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    X_res, y_res = SMOTE(random_state=42).fit_resample(X_train, y_train)

    model = LGBMClassifier(**mejores_params)
    model.fit(X_res, y_res)

    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    resultados.append([
        accuracy_score(y_test, y_pred),
        precision_score(y_test, y_pred),
        recall_score(y_test, y_pred),
        f1_score(y_test, y_pred)
    ])
    loglosses.append(log_loss(y_test, y_prob))
    aucs.append(roc_auc_score(y_test, y_prob))

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    precision, recall, _ = precision_recall_curve(y_test, y_prob)
    roc_curves.append((fpr, tpr))
    prc_curves.append((recall, precision))

    y_true_total.extend(y_test)
    y_prob_total.extend(y_prob)

    del model, X_train, X_test, y_train, y_test, y_pred, y_prob, X_res, y_res
    gc.collect()

# ðŸ“ˆ MÃ©tricas
res_np = np.array(resultados)
metrics_mean = res_np.mean(axis=0)
metrics_std = res_np.std(axis=0)
logl_mean, logl_std = np.mean(loglosses), np.std(loglosses)
auc_mean, auc_std = np.mean(aucs), np.std(aucs)

# ðŸ“ˆ GrÃ¡fica ROC
fpr_interp = np.linspace(0, 1, 100)
tprs = [np.interp(fpr_interp, fpr, tpr) for fpr, tpr in roc_curves]
tpr_mean, tpr_std = np.mean(tprs, axis=0), np.std(tprs, axis=0)

plt.figure()
plt.plot(fpr_interp, tpr_mean, label=f"AUC = {auc_mean:.3f}")
plt.fill_between(fpr_interp, tpr_mean - tpr_std, tpr_mean + tpr_std, alpha=0.2)
plt.plot([0,1],[0,1],'k--')
plt.title("Curva ROC promedio - LightGBM (Optuna + SMOTE)")
plt.xlabel("Falsos Positivos")
plt.ylabel("Verdaderos Positivos")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig(ruta_roc)
plt.show()

# ðŸ“ˆ GrÃ¡fica PRC
recall_interp = np.linspace(0, 1, 100)
precisions = [np.interp(recall_interp, r[::-1], p[::-1]) for r, p in prc_curves]
prec_mean, prec_std = np.mean(precisions, axis=0), np.std(precisions, axis=0)

plt.figure()
plt.plot(recall_interp, prec_mean, label=f"AP = {average_precision_score(y_true_total, y_prob_total):.3f}")
plt.fill_between(recall_interp, prec_mean - prec_std, prec_mean + prec_std, alpha=0.2)
plt.title("Curva PRC promedio - LightGBM (Optuna + SMOTE)")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig(ruta_prc)
plt.show()

# ðŸ§¾ Guardar resumen
if os.path.exists(ruta_resumen):
    resumen = pd.read_csv(ruta_resumen)
    if 'ObservaciÃ³n' in resumen.columns:
        resumen = resumen.drop(columns='ObservaciÃ³n')
else:
    resumen = pd.DataFrame(columns=[
        'Base', 'Modelo', 'Naturaleza', 'Temporalidad', 'Tipo de aprendizaje',
        'Accuracy', 'Desv. Accuracy', 'Precision', 'Desv. Precision',
        'Recall', 'Desv. Recall', 'F1-score promedio', 'DesviaciÃ³n F1',
        'AUC', 'Desv. AUC', 'LogLoss', 'Desv. LogLoss',
        'Top 1 variable', 'Top 2 variable', 'Top 3 variable'
    ])

idx = resumen[
    (resumen['Base'] == 'Turismo') &
    (resumen['Modelo'] == 'LightGBM (Optuna + SMOTE)')
].index
idx = idx[0] if len(idx) > 0 else len(resumen)

# ðŸ“Œ Top 3 variables
model_final = LGBMClassifier(**mejores_params)
model_final.fit(X, y)
importancias = pd.Series(model_final.feature_importances_, index=X.columns)
top_vars = importancias.sort_values(ascending=False).head(3).index.tolist()
top1, top2, top3 = (top_vars + [None]*3)[:3]

resumen.loc[idx] = [
    'Turismo', 'LightGBM (Optuna + SMOTE)', 'Avanzado', 'EstÃ¡tica', 'Supervisado',
    round(metrics_mean[0], 4), round(metrics_std[0], 4),
    round(metrics_mean[1], 4), round(metrics_std[1], 4),
    round(metrics_mean[2], 4), round(metrics_std[2], 4),
    round(metrics_mean[3], 4), f"Â±{metrics_std[3]:.4f}",
    round(auc_mean, 4), round(auc_std, 4),
    round(logl_mean, 4), round(logl_std, 4),
    top1, top2, top3
]

resumen.to_csv(ruta_resumen, index=False)
print(f"âœ… Resultados guardados en: {ruta_resumen}")
```
